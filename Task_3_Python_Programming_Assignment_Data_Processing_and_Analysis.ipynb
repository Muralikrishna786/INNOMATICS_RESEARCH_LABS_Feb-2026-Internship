```python
#Employee Performance Bonus Eligibility
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}

#Find highest score
highest_score = max(employees.values())

#Find all employees with highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", ", ".join(top_performers),f"(Score: {highest_score})")
```

```python
#Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
for char in string.punctuation:
    query = query.replace(char, "")
words = query.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display words searched more than once
result = {}
for word, count in frequency.items():
    if count > 1:
        result[word] = count
print(result)
```

```python
#Sensor Data Validation
sensor_readings=[3,4,7,8,10,12,5]
valid_readings=[]
for hour_index,value in enumerate(sensor_readings):
    if value%2==0:
        valid_readings.append((hour_index,value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Email Domain Usage Analysis

#Input Emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
#Find Domain Count
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain]=domain_count.get(domain,0)+1
total = len(emails)
for domain,count in domain_count.items():
    percentage=(count/total)*100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
#Find Average
average = sum(sales) / len(sales)

# Modified threshold (approx 30% increase)
threshold = average + 500
for day, value in enumerate(sales, start=1):
    if value >= threshold:
        print(f"Day {day}: {value}")
```

```python
#Duplicate User ID Detection

user_ids=["user1","user2","user1","user3","user1","user3"]

#Find the Frequency of Each Duplicate Id's
frequency={}
for user in user_ids:
    frequency[user]=frequency.get(user,0)+1
for user,count in frequency.items():
    if count>1:
        print(f"{user} → {count} times")
```
