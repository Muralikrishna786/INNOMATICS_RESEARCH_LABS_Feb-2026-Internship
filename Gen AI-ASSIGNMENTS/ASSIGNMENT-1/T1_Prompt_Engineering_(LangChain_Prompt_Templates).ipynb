```python
# Replace Hardcoded Prompts
def get_explain_prompt():
    template = "Explain {topic} in simple terms for beginners"

    prompt = PromptTemplate(
        input_variables=["topic"],
        template=template
    )

    return prompt

# Test
prompt1 = get_explain_prompt()
print(prompt1.format(topic="Artificial Intelligence"))
```

```python
# Multi - Input Prompts
def get_multi_input_prompt():
    template = "Explain {topic} for {audience} in a {tone} tone"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt

# Test Cases
multi_prompt = get_multi_input_prompt()

print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# Prompt Variations Engine
def get_prompt_variations():

    teaching_template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    )

    interview_template = PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    )

    storytelling_template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    )

    return teaching_template, interview_template, storytelling_template


# Run
topic = "Machine Learning"

teach, interview, story = get_prompt_variations()

print(teach.format(topic=topic))
print(interview.format(topic=topic))
print(story.format(topic=topic))
```

```python
# Chat Prompt Template System
def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking insightful questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker who inspires learners."
    else:
        system_msg = "You are a helpful assistant."

    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Tell me about {topic}")
    ])

    return chat_prompt

# Example
chat = get_chat_prompt("teacher", "Neural Networks")
print(chat.format_messages(topic="Neural Networks"))
```

```python
# Input Validation layer
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("kids", "friendly"))  # Defaults applied
```

```python
# Prompt Generation APP
def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Define styles
    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for a {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    # Step 3: Select template
    template = templates.get(style, templates["teaching"])

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)


# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# Template Reusability Test
def reusable_template_test():

    template = "Explain {topic} for {audience} in a {tone} tone"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    test_cases = [
        {"topic": "AI", "audience": "beginner", "tone": "fun"},
        {"topic": "Python", "audience": "intermediate", "tone": "formal"},
        {"topic": "Data Science", "audience": "expert", "tone": "casual"},
        {"topic": "Machine Learning", "audience": "beginner", "tone": "formal"},
        {"topic": "Deep Learning", "audience": "expert", "tone": "fun"}
    ]

    for case in test_cases:
        print(prompt.format(**case))


# Run
reusable_template_test()
```

```python
# Final Pipeline Flow
def prompt_pipeline(topic, audience, tone, style):

    # Step 1: Validate
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Generate prompt
    result = generate_prompt(topic, audience, tone, style)

    return result


# Example
print(prompt_pipeline("Transformers", "beginner", "fun", "teaching"))
```
