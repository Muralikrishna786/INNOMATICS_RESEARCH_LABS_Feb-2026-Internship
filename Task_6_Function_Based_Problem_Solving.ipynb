```python
# Online Food Delivery Time Estimator
def delivery_time_estimator(distance,wheather,traffic):
  eta=distance*5
  if traffic=="high":
    eta+=15
  elif traffic=="moderate":
    eta+=8
  if wheather=="rainy":
    eta+=10
  elif wheather=="strom":
    eta+=20
  print("Estimated Delivery Time:",int(eta),"minutes")
# inputs
distance=float(input("Enter the Distance(km):"))
Traffic=input("Enter the Traffic(High/Moderate):").lower()
wheather=input("Enter the Wheather(Rainy/Stormy):").lower()
delivery_time_estimator(distance,wheather,Traffic)
```

```python
# Movie Theatre Seat Occupancy Analyzer
def seat_occupancy_analyzer(total_seats,booked_seats):
  booked=len(booked_seats)
  occupancy_percentage=(booked/total_seats)*100
  print("Occupancy:",int(occupancy_percentage),"%")

  if(occupancy_percentage==100):
    status="House Full"
    print("Suggestion: Opening Additional Shows")
  elif(occupancy_percentage>=75):
    status="Almost Full"
  else:
    status="Seats Available"
  print("Show Status:",status)

#inputs
Total_seats=200
Booked_seats=[1]*150
seat_occupancy_analyzer(Total_seats,Booked_seats)

```

```python
# Cloud Server Load Classification System
def load_classification(CPU_Readings):
  average=sum(CPU_Readings)/len(CPU_Readings)
  print("Average CPU Load:",average,"%")

  if average<50:
    status="Normal"
  elif average>=50 and average<80:
    status="Warning"
  else:
    status="Critical"
  print("Server Status:",status)

CPU_Readings=[45,60,70,85,90]
load_classification(CPU_Readings)
```

```python
# Smart Classroom Resource Usage Monitor
def resource_monitor(resources):
    overused = []
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
# Online Event Registration Capacity Controller
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

# Sample Input
event_registration(100, 105)
```
