Task 1: Conceptual Understanding
1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are considered different tokens if text is case-sensitive.
This means a model may treat them as two separate words even though they have the same meaning.

Example:

"I Love NLP"
"I love NLP"

Tokens without lowercasing:

["I", "Love", "NLP"]
["I", "love", "NLP"]

After lowercasing:

["i", "love", "nlp"]
["i", "love", "nlp"]

So lowercasing helps reduce duplicate vocabulary and improves model performance.

2. What happens if stopwords are not removed?

If stopwords are not removed:

Text contains many unnecessary words like the, is, at, on
Increases dataset size
Slows processing
Important words get less importance
Models may learn noise instead of meaning

However, sometimes stopwords are important (see next question).

3. Real-world scenarios where removing stopwords is harmful
Scenario 1: Sentiment Analysis

Sentence:

"I am not happy"

If stopwords removed:

happy

Meaning becomes positive instead of negative.

Scenario 2: Chatbots / Question Answering

Sentence:

"What is the time now?"

Removing stopwords may break sentence structure and meaning.

So stopwords removal must be task-dependent.

4. Difference between Stemming and Lemmatization
Stemming	Lemmatization
Cuts words to root form	Converts word to dictionary form
Faster	Slower
May produce non-words	Produces valid words
Example: "running" → "run"	"running" → "run"
Example: "studies" → "studi"	"studies" → "study"

Summary:

Stemming = crude cutting
Lemmatization = meaningful root word

```python
#Task-2: Build Advanced Preprocessing Function

import re

def preprocess_text(text):
    try:
        if not text or text.strip() == "":
            return [], ""

        # Remove URLs and emails
        text = re.sub(r'http\S+|www\S+|https\S+', '', text)
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Convert to lowercase
        text = text.lower()

        # Handle repeated characters (soooo -> so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove special characters / emojis
        text = re.sub(r'[^a-z\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenization
        tokens = text.split()

        # Remove very short tokens except no, not
        filtered_tokens = []
        for word in tokens:
            if len(word) > 2 or word in ["no", "not"]:
                filtered_tokens.append(word)

        cleaned_sentence = " ".join(filtered_tokens)

        return filtered_tokens, cleaned_sentence

    except Exception as e:
        return [], ""
```

```python
#Task-3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product @!!#",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print()
```

```python
#Task-4: Token Analytics

def token_analytics(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    total, unique, avg = token_analytics(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print()
```

```python
#Task-5: Frequency Analysis

from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(word_freq.most_common(10))

print("\nLeast 5 Frequent Words:")
print(word_freq.most_common()[-5:])
```

```python
#Task-6: Build Full Pipeline

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)
print(result)
```

```python
#Task-7: Error Handling

edge_cases = [
    "",
    "!@#$%",
    "1234567890"
]

for text in edge_cases:
    tokens, clean = preprocess_text(text)
    print("Input:", text)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print()
```
