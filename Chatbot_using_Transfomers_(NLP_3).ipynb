1. Introduction

A chatbot is a conversational AI system that interacts with users in natural language. Traditional chatbots used rule-based systems, but modern chatbots use Transformer-based models like GPT and DialoGPT to generate intelligent responses.

Transformers are deep learning models designed for Natural Language Processing tasks such as:

Text generation
Translation
Question answering
Chatbots

In this assignment, we build a console-based chatbot using a pre-trained model from Hugging Face Transformers.

2. Pipeline Flow

Chatbot Pipeline

User Input → Tokenization → Transformer Model → Response Generation → Output → Loop


Step-by-Step Flow

1) User enters text
2) Text is converted into tokens
3) Tokens are given to transformer model
4) Model generates response
5) Response converted back to text
6) Printed to console
7) Loop continues until user types exit

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
print("Chatbot: Hello! I am your AI assistant. Type 'exit' or 'quit' to end the chat.")

chat_history = ""

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break

    # Create prompt
    prompt = chat_history + "User: " + user_input + "\nChatbot:"

    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=150,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.9
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract chatbot response only
    response = response.split("Chatbot:")[-1].strip()

    print("Chatbot:", response)

    # Update chat history
    chat_history += "User: " + user_input + "\nChatbot: " + response + "\n"
```
