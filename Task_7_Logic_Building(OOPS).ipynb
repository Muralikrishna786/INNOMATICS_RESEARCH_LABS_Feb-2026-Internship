```python
# Smart Light Controller
class smart_light:
  def __init__(self,light_name):
    self.light_name = light_name
    self.light_status = "OFF"

  def light_on(self):
    self.light_status = "ON"
  def light_off(self):
    self.light_status = "OFF"

  def diaplay_status(self):
    print(f"{self.light_name} light is {self.light_status}")

# Object Creation
light=smart_light("Bed Room")
light.light_on()
light.diaplay_status()
```

```python
# Employee ID Card System
class employee:
  def __init__(self,emp_name,emp_id,emp_dept):
    self.emp_name=emp_name
    self.emp_id=emp_id
    self.emp_dept=emp_dept

  def dispaly(self):
    print("Employee ID Card")
    print(f"Name:{self.emp_name}")
    print(f"ID:{self.emp_id}")
    print(f"Department:{self.emp_dept}")

emp1=employee("Rahul","EMP102","IT")
emp1.dispaly()
```

```python
# Mobile Contact Record
class contacts:
  def __init__(self,name,phone_number):
    self.name=name
    self.phone_number=phone_number

  def dispaly(self):
    print("Contact Saved")
    print(f"Name:{self.name}")
    print(f"Phone:{self.phone_number}")

contact1=contacts("Anita",9876543210)
contact1.dispaly()
```

```python
# Product Price Tag Generator
class products:
  def __init__(self,p_name,p_price):
    self.name = p_name
    self.price = p_price
  def display(self):
    print(f"Product Name:{self.name}")
    print(f"Price:{self.price}")

product1=products("Headphones",2499)
product1.display()
```

```python
# Movie Rating Display System
class movies:
  def __init__(self,movie_name,rating):
    self.movie_name=movie_name
    self.rating=rating
  def display(self):
    print(f"Movie Name:{self.movie_name}")
    print(f"Rating:{self.rating} / 5")

movie1=movies("Inception",4.8)
movie1.display()
```

```python
# Delivery Address Manager
class delivery_details:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  def dispaly(self):
    print("Delivery Details")
    print("Customer Name:",self.name)
    print("Address:",self.address)

customer1=delivery_details("Suman","Hyderabad")
customer1.dispaly()
```
