```python
#Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.\n")
            continue

        if amount in valid_plans:
            print("Recharge Successful ✅")
            break
        else:
            print("Invalid Plan. Available plans: 199, 299, 399, 599\n")

# Call function
validate_recharge()
```

```python
#Inventory Reorder Alert System
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item} → Reorder Alert")
        else:
            print(f"{item} → Stock OK")

# Example dictionary
products = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 30,
    "Salt": 5
}
check_inventory(products)
```

```python
#Student Result Processing System
def result_processing(marks):
  total=0
  for mark in marks:
    total+=mark
  average=total/len(marks)
  print("Average Marks:",average)
  if (average>=50):
    return "Pass"
  else:
    return "Fail"
#Example
student_marks=[15,30,78,22,18]
result=result_processing(student_marks)
print("Result:",result)
```

```python
#Cab Fare Estimator with Retry Option
def calculate_cab_fare(distance,peak):
  base_fare=50
  fare_per_km=12
  fare=base_fare+(fare_per_km*distance)
  if peak.lower()=="yes":
    fare=fare + (fare*0.25)
  return fare
while True:
  km=float(input("Enter distance in km: "))
  peak=input("Is it a peak hour? (yes/no): ")
  total_fare=calculate_cab_fare(km,peak)
  print("Total Fare: ₹",total_fare)
  retry=input("Do you want to calculate fare again? (yes/no): ")
  if retry.lower()!="yes":
    break
```

```python
#Employee Attendance Eligibility
def check_attendance(attendance_list):
  present_days=0
  for attendance in attendance_list:
    if attendance=='P':
      present_days+=1
  percentage=(present_days/len(attendance_list))*100
  print("Attendance Percentage is:",percentage)
  if (percentage>=75):
    return "Eligible"
  else:
    return "Not Eligible"
#Example
attendence_list=['P','P','P','A','P','P','A','P','A','P']
status=check_attendance(attendence_list)
print("Status:",status)
```
