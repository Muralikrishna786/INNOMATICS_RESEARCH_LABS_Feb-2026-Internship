```python
#Social Media - Post Engagement Analyzer
def engagement_Analyze(likes):
  total_likes=0
  for like in likes:
    total_likes+=like
  if total_likes>=100:
    status="Viral Post"
  else:
    status="Normal Engagement"
  print("Total Likes:",total_likes)
  print("Post Status:",status)

likes_list=[200,300,250,300]
engagement_Analyze(likes_list)
```

```python
# Agriculture - Rainfall Adequacy Checker
def rainfall_check(rainfall,required_level):
  total_rainfall=0
  for rain in rainfall:
    total_rainfall+=rain
  average_rainfall=total_rainfall//len(rainfall)
  if (average_rainfall>=required_level):
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print("Average Rainfall:",average_rainfall)
  print("Rainfall Status:",status)

rainfall_list=[70,75,80,63]
rainfall_check(rainfall_list,70)
```

```python
# Social Media - Duplicate Account Detection
def account_detection(usernames):
  unique_users=set(usernames)
  if len(unique_users)==len(usernames):
    result="No"
  else:
    result="YES"
  print("Duplicate Accounts Found:",result)

usernames_list=["user1","user2","user3","user1"]
account_detection(usernames_list)
```

```python
# Agriculture - Premium Crop Price Filter
def filter_crops(prices):
  premium=[]
  for price in prices:
    if (price>2000):
      premium.append(price)
  print("Premium Crops:",premium)

crop_prices=[1500,2500,1800,3200]
filter_crops(crop_prices)
```

```python
# System Monitoring - Application Health Checker
def application_check(errors):
  if(errors==0):
    status="Healthy"
  elif(errors<=5):
    status="Minor Issues"
  else:
    status="Critical Issues"
  print("Error Count:",errors)
  print("System Status:",status)
errors_count=int(input("Enter Error Count:"))
application_check(errors_count)
```

```python
# E-Learning - Student Attendance Eligibility System
def attendance_eligibility(attendance_list):
  total_classes=len(attendance_list)
  attended=0
  for attendance in attendance_list:
    if(attendance==1):
      attended+=1
  percentage=(attended/total_classes)*100
  if(percentage>=75):
    status="Eligible"
  else:
    status="Not Eligible"
  print("Attendance Percentage:",percentage)
  print("Exam Eligibility:",status)

attendance_list=[1,1,1,0,1]
attendance_eligibility(attendance_list)
```

```python
# Smart Electicity Bill Analyzer
def calculate_bill(units):
  bill=0
  if(units<=100):
    bill=units*3
  elif(units>100 and units<=200):
    bill=100*3+(units-100)*5
  else:
    bill=100*3+100*5+(units-100)*7

  if (bill<500):
    print("Low Usage")
  elif(bill>=500 and bill<=1500):
    print("Moderate Usage")
  else:
    print("High Usage")

units=int(input("Enter the units consumed:"))
calculate_bill(units)
```
