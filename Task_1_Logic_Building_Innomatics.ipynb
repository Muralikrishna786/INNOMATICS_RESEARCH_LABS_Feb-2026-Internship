```python
#User Login Check
#Inputs
username_input = input("Enter username: ")
password_input = input("Enter password: ")
username_input = username_input.lower()
username= "admin"
password= "1234"

#Check Wheather the credentials are matched or not
if(username_input==username and password_input==password):
  print("Login Successfull")
else:
  print("Invalid Credentials")
```

```python
#Pass/Fail Analyzer

#Inputs
marks=[45,78,90,33,60]
pass_students=0
fail_students=0

#Automation use for loop
for i in marks:
  if i>=50:
    pass_students+=1
  else:
    fail_students+=1
print("Pass Students: ",pass_students)
print("Fail Students: ",fail_students)
```

```python
#Simple Data Cleaner

names=["Alice ","bob"," CHARLIE "]
cleaned_names=[]
for i in names:
  cleaned_name=i.strip().lower()
  cleaned_names.append(cleaned_name)
print("Cleaned List:",cleaned_names)
```

```python
#Message Length Analyzer

messages=["Hi","Welcome to the platform","OK"]

for i in messages:
  message_length=len(i)
  print(f"Message:{i} | Length:{message_length}")
  if message_length>10:
    print("Flag:Messages is longer-than 10 characters")
```

```python
#Error Message Detector

logs=["INFO","ERROR","WARNING","ERROR"]

#Error text counts by using count
error_count=logs.count("ERROR")
print("Total Error Count:",error_count)
```
